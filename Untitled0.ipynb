```python
# PD.SERIES (SIMILAR TO DICTIONARY)
```

```python
import pandas as pd

data = pd.Series([3, 0.3, 4.5, 99, 7.9], index = ['a', 'b', 'c', 'd', 'e'])
(data.values)
(data.index)
data["c"]
data["a": "d"]
```

```python
# THIS "name_dict" IS A DICTIONARY
anime_dict = {'a': "naruto", 'b': "sasuke", 'c': "sakura", 'd': "kakashi"}
style_dict = {'a': "wind", 'b': "fire", 'c': "medic", 'd': "lightning"}
style = pd.Series(style_dict)
anime = pd.Series(anime_dict)
anime.values
anime.index
anime['b']
anime[::-1]
anime['a':: 2]
anime['b': 'd']
anime
style
```

```python
# PD.DATAFRAMES
```

```python
nama = pd.DataFrame({'anime': anime, 'style': style})

print(nama)
nama.T
nama
nama.values[3, 0]
nama.columns
```
