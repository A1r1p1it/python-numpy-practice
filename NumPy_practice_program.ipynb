```python
#Problem 8: Sales Filter
#Create a "sales" array with 50 random values (100-1000). Find:
#- All sales above 500
#- Sales between 300-700
#x- Replace all sales below 200 with 200 (minimum guarantee)
```

```python
import numpy as np
arr = np.random.randint(100, 1001, 50)
print("50 random values:\n", arr, "\n")

sal_500 = arr[arr > 500]
print("All sales above 500:\n", sal_500, "\n")

sal_300_700 = arr[(arr > 300) & (arr < 700)]
print("Sales between 300-700: ", sal_300_700, "\n")

arr[arr < 200] = 200
print("all sales below 200 with 200: ", arr,"\n")
```

```python
### Problem 9: Multi-Condition Filter
#Generate 100 random numbers (1-100). Find numbers that are:
#- Greater than 50 AND divisible by 3
#- Less than 30 OR greater than 80
#- Between 40-60 AND even
```

```python
arr = np.random.randint(1, 101, 100)
print("100 random numbers: \n", arr, "\n")

divide_3 = arr[(arr > 50) & (arr % 3 == 0)]
print("Greater than 50 AND divisible by 3: \n", divide_3, "\n")

less_30 = arr[(arr < 30) | (arr > 80)]
print("Less than 30 OR greater than 80: \n", less_30, "\n")

bet_40_60 = arr[(arr > 40) & (arr < 60) & (arr % 2 == 0)]
print("Between 40-60 AND even: \n", bet_40_60, "\n")

######################### or is typed as '|' DOUBTTTTTTTTTTTTTTTTTTTTTT
```

```python
### Problem 11: Random Walk Simulator
#- Start with position 0
#- Generate 1000 random steps: each step is +1 or -1 (use randint)
#- Use cumulative sum to track position over time
#- Find the maximum and minimum positions reached
```

```python
arr = np.random.randint(0, 2, 1000)
arr[arr == 0] = -1
#arr = np.random.choice([-1, 1], size=1000)
print("Generates 1000 steps: ", arr, "\n")

cum_pos = np.cumsum(arr)
print("cumulative sum to track position over time: ", cum_pos, "\n")

max = np.max(cum_pos)
print("maximum position: ", max,"\n")

min = np.min(cum_pos)
print("minimum position: ", min,"\n")
```

```python
import numpy as np

# Create array with 100 random numbers (1-100)
arr = np.random.randint(1, 101, 100)
print("Original array (first 20):", arr[:20], "\n")

# Randomly replace 10 positions with 999
random_indices = np.random.choice(100, size=10, replace=False)
arr[random_indices] = 999
print("After adding 999s (first 20):", arr[:20], "\n")

# Use masking to find all 999 values
mask = arr == 999
print("Mask for 999 values:", mask, "\n")
print("Number of 999s:", np.sum(mask), "\n")

# Calculate mean (excluding 999s for accurate mean)
mean = np.mean(arr[arr != 999])
print("Array's mean (excluding 999s):", mean, "\n")

# Replace 999s with mean
arr[mask] = mean
print("After replacing 999s with mean (first 20):", arr[:20], "\n")

# Find how many values are above the new mean
new_mean = np.mean(arr)
above_mean_count = np.sum(arr > new_mean)
print("New mean:", new_mean)
print("Number of values above new mean:", above_mean_count)
```
